```python
mydataset=pd.read_csv("train.csv")
```

```python
from pycaret.regression import *
setup(data=mydataset,target='yield',feature_selection=True,normalize=True,normalize_method="minmax",transformation=True,outliers_method="0.1",train_size=0.8)#pca#transformation=True
```

```python
cm=compare_models()
```

```python
mydataset.corr()['yield'].sort_values(ascending=False)
```

```python
lightgbmModel=create_model("gbr")
```

```python
mynewdata=pd.read_csv("test.csv")
```

```python
predictions=predict_model(lightgbmModel,data=mynewdata)
predictions
```

```python
dictionary=pd.DataFrame({'id':predictions['id'],'yield':predictions['prediction_label']})
dictionary
```

```python
dictionary.to_csv("Answer.csv",index=False)
```
